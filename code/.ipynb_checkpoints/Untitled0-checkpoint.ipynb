```python
import pandas as pd
```

```python
diag = pd.read_csv("/Users/kanc/Documents/classes/Kaggle/practice_fusion/data/trainingSet/training_SyncDiagnosis.csv")
pat = pd.read_csv("/Users/kanc/Documents/classes/Kaggle/practice_fusion/data/trainingSet/training_SyncPatient.csv")
pres = pd.read_csv("/Users/kanc/Documents/classes/Kaggle/practice_fusion/data/trainingSet/training_SyncPrescription.csv")
lab_result = pd.read_csv("/Users/kanc/Documents/classes/Kaggle/practice_fusion/data/trainingSet/training_SyncLabResult.csv")
lab_panel = pd.read_csv("/Users/kanc/Documents/classes/Kaggle/practice_fusion/data/trainingSet/training_SyncLabPanel.csv")
lab_obs = pd.read_csv("/Users/kanc/Documents/classes/Kaggle/practice_fusion/data/trainingSet/training_SyncLabObservation.csv")
trans = pd.read_csv("/Users/kanc/Documents/classes/Kaggle/practice_fusion/data/trainingSet/training_SyncTranscript.csv")
```

```python
all_labs = lab_panel.merge(lab_obs, on='LabPanelGuid').merge(lab_result, on='LabResultGuid')
all_labs.HL7Text.value_counts()
```

```python
pat.merge(trans,on='PatientGuid').groupby('PatientGuid').first()
```

```python
simplefeatures = trans.groupby('PatientGuid').first()
simplefeatures = simplefeatures[[u'Height', u'Weight', u'BMI', u'SystolicBP', u'DiastolicBP']]
simplefeatures.dropna()
len(simplefeatures)
```

```python
diag_pat = diag.merge(pat, on='PatientGuid')
all_labs[diag_pat.ICD9Code.str.contains('272')].HL7Text.value_counts()
                                    
```

```python
results = pat.merge(pres,on='PatientGuid').groupby('PatientGuid').size()
pat_results = pat.join(results.to_frame(), on='PatientGuid')
pat_results.columns = [u'PatientGuid', u'Gender', u'YearOfBirth', u'State', u'PracticeGuid', u'prescription_count']
```

```python
patfeat_results = pat_results.join(simplefeatures, on='PatientGuid')
```

```python
# dropna

patfeat_results = patfeat_results.dropna()
patfeat_results


# modifying gender
patfeat_results.Gender.value_counts()
patfeat_results['Gender'] = patfeat_results['Gender'].map( {'F': 0, 'M': 1} ).astype(int)
patfeat_results
```

```python
from sklearn.cross_validation import train_test_split
```

```python
train, test = train_test_split(patfeat_results, test_size=0.5, random_state=5)
patfeat_results.columns

```

```python
traindf = pd.DataFrame(train, columns = patfeat_results.columns)
testdf = pd.DataFrame(test, columns = patfeat_results.columns)
```
